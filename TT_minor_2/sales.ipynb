#SALES ANALYSIS 

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('Stores.csv')
df["monthly sales"]=df["Daily_Customer_Count"]*30
df["Average sales"]=df["Store_Sales"]/30
sort=df.sort_values(by=["monthly sales"], ascending=False)
print(sort.head(10))
maxi=np.max(df["monthly sales"])
print("Maximum Monthly Customer Count:",maxi)

mini=np.min(df["monthly sales"])
print("Minimum Monthly Customer Count:",mini)   
mean=np.mean(df["monthly sales"])
print("Mean Monthly Customer Count:",mean)
plt.hist(df["monthly sales"], bins=10, color='blue', edgecolor='black')
plt.title('Histogram of Monthly Customer Count')
plt.xlabel('Monthly Customer Count')
plt.ylabel('Rate of sales ')

plt.figure(figsize=(8,6))

plt.figure(figsize=(8,6))
plt.scatter(df["monthly sales"], df["Average sales"], color='red')
plt.title("Monthly Sales vs Average Sales")
plt.xlabel("Monthly Customer Count")
plt.ylabel("Average Sales")


plt.figure(figsize=(8,6))
plt.scatter(df["Store_Area"], df["Daily_Customer_Count"], color='purple', edgecolor='black')
plt.title("Store Area vs Daily Customer Count")
plt.xlabel("Store Area (sq ft)")
plt.ylabel("Daily Customer Count")

df_sorted_area = df.sort_values(by="Store_Area")
plt.figure(figsize=(8,6))
plt.plot(df_sorted_area["Store_Area"], df_sorted_area["Daily_Customer_Count"], marker='o', linestyle='-', color='blue')
plt.title("Trend: Store Area vs Daily Customer Count")
plt.xlabel("Store Area (sq ft)")
plt.ylabel("Daily Customer Count")
plt.show()
```
